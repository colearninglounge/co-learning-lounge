### Slicing and Indexing Data

```
# .loc[] is a widely used function for slicing through dataframes

# Suppose we want to slice the dataframe through the gender column 

male = data_2.loc[data_2['gender'] == 'Male']
male.head()
```

```
# Further we can also slice a dataframe into another dataframe with only required columns

new_data_2 = data_2.loc[:,['customerID','gender','Dependents']]
new_data_2.head()
```

The above code would give the specified columns and all the rows. We can slice both over rows and columns simultaneously as well.

```
# Slicing the dataframe to a new dataframe with first 100 rows and first 3 columns

new_data_3 = data_2.loc[:100,['customerID','gender','Dependents']]
new_data_3.tail()
```

Another function similar to .loc[] is the .iloc[] function ('i' stands for index), since i iterates over index.

```
# Slicing the dataframe to a new dataframe with first 100 rows and first 3 columns using .iloc[] function

new_data_3_1 = data_2.iloc[:100,:3]
new_data_3_1.tail()
```

```
new_data_3_2 = data_2.iloc[:100,4:6]
new_data_3_2.tail()
```

```
new_data_3_3 = data_2.iloc[:100,3:5]
new_data_3_3.tail()
```

 Note : .iloc[] would iterate over the index and so it will result in the first 100 rows of the dataframe unlike .loc[] which would output upto the 100th row of the dataframe.

### The groupby() and filter() function


The filter() function subsets the dataframe according to the mentioned labels.

```
# Filtering the dataframe according to the mentioned items.
# The labels mentioned in the items paramater should be the columns and so one doesn't need to
# mention the axis for the same

data_2.filter(items = ['tenure', 'PhoneService'])
```

```
# Using the like parameter to filter the dataframe
# The axis = 1 states that filtering is performed across columns and all the columns
# with the characters "Charge" in them will constitute the output

data_2.filter(like = 'Charge', axis = 1)
```

```
# We can also filter the dataframe based on index. 
# Set the 'gender' column as the index
data_2.set_index('gender', inplace = True)

# Filter the index using like parameter and axis = 0 states filtering is across rows
# This will output all the rows with index = Female
data_2.filter(like = 'Female', axis = 0)
```

```
data_2.reset_index(inplace = True)
```

```
# Grouping the data by gender and calculating the mean monthly charges gender wise

data_2.groupby(['gender'])['MonthlyCharges'].mean()
```

```
# Multiple Grouping
# Grouping the data first by Gender and then by Dependents and calculates average Monthly Charges
# paid by all the categories
data_2.groupby(['gender','Dependents'])['MonthlyCharges'].mean()
```

```
# Setting up a custom index
# We can set up a column as an index

data_2.set_index(['customerID'], inplace=True)
```

```
data_2.head()
```

```
# we can reset the index back to the default state 

data_2.reset_index(inplace = True)
```

### Merging two dataframes

```
# Joining the two dataframes and performing the join on the index, both the dataframes have a common index

new_data_merge = pd.merge(new_data_3_1, new_data_3_2, left_index=True, right_index=True)
new_data_merge
```

```
# The concat method performs the same function as the merge function above 
# Typical SQL type outer join
# axis parameter set to 1 indicates both dataframes are merged/joined n columns
new_data_concat = pd.concat(objs = [new_data_3_1,new_data_3_2], axis = 1)
new_data_concat
```

```
# Suppose, we want to merge to dataframes on a common column among both the dataframes
# Merging dataframes on the 'PhoneService' column and performing a SQL type left join
new_data_3_left = pd.merge(new_data_3_3, new_data_3_2, on='PhoneService', how='left')
new_data_3_left
```

```
# Using concat function to perform a SQL type inner join i.e. intersection or
# common columns among both the dataframes

new_data_3_inner = pd.concat(objs = [new_data_3_3, new_data_3_2], join='inner')
new_data_3_inner
```
